# ML

#### When new restaurant enters Top 50 world's restaurants, regarding its location, price tier, rating, and type of kitchen - will it have 3 Michelin stars?

##### I am aware of the fact, that the dataset is small, and whether the restaurant gets 3 stars does not solely depends on the attributes listed above - so this ML is run just out of curiosity how successful the algorithms will be.

```python
df_train=df1.copy()
```

```python
df_train.head()
```

```python
df_countries=pd.get_dummies(df_train["Country"],drop_first=True)
```

```python
df_type=pd.get_dummies(df_train["Category"],drop_first=True)
```

```python
#df_train=pd.concat([df_train,df_countries,df_type],axis=1)
```

```python
X=pd.concat([df_countries,df_type,df_train.drop(["Has 3 Michelin stars","Country","City","Category","ID_fsq","Name_fsq","Name","Price_Category","Rating","Lat","Lng"],1)],axis=1)
```

```python
X.shape
```

```python
X.columns
```

```python
y=df_train.loc[:,"Has 3 Michelin stars"].astype(int)
```

```python
#y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn.fit(X_train,y_train)
```

```python
y_pred=knn.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
knn.fit(X,y)
```

```python
y_pred4=knn.predict(X)
```

```python
print(confusion_matrix(y,y_pred4))
print(classification_report(y,y_pred4))
print(accuracy_score(y,y_pred4))
```
