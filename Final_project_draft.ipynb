# ML

```python
df_train=df1.copy()
```

```python
df_countries=pd.get_dummies(df_train["Country"],drop_first=True)
```

```python
df_type=pd.get_dummies(df_train["Category"],drop_first=True)
```

```python
df_train=pd.concat([df_train,df_countries,df_type],axis=1)
```

```python
X=pd.concat([df_countries,df_type],axis=1)
```

```python
X.shape
```

```python
y=df_train.loc[:,"Has 3 Michelin stars"].astype(int)
```

```python
#y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25,random_state=0)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans=KMeans(n_clusters=2)
```

```python
kmeans.fit(X_train,y_train)
```

```python
y_pred=kmeans.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
kmeans.fit(X,y)
```

```python
y_pred4=kmeans.predict(X)
```

```python
print(confusion_matrix(y,y_pred4))
print(classification_report(y,y_pred4))
print(accuracy_score(y,y_pred4))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred2=lr.predict(X_test)
```

```python
print(confusion_matrix(y_test,y_pred2))
print(classification_report(y_test,y_pred2))
print(accuracy_score(y_test,y_pred2))
```
